### 16. 두 수의 덧셈

역순으로 저장된 연결리스트의 숫자를 더하라

1 -> 2 -> 3

5 -> 6 -> 7

321 + 765 = 1086 이 결과를 계산하고 이것을 다시 역순으로 리턴한다.

6 -> 8 -> 0 -> 1

주의! 123 + 567 = 690 이 아니다.

```
# 연결 리스트 뒤집기
def reverseList(self, head: ListNode) -> ListNode: # 역순 연결리스트를 생성합니다. 설명 생략(사진 O)
    node, prev = head, None

    while node:
        next, node.next = node.next, prev
        prev, node = node, next

    return prev

# 연결 리스트를 파이썬 리스트로 변환
def toList(self, node: ListNode) -> List: # .next로 하나씩 불러가며 빈 리스트에 추가한다.
    list: List = []
    while node:
        list.append(node.val)
        node = node.next
    return list

# 파이썬 리스트를 연결 리스트로 변환
def toReversedLinkedList(self, result: str) -> ListNode: # 리스트를 받아서 역순으로 연결리스트를 생성한다.
    prev: ListNode = None
    for r in result:
        node = ListNode(r) # 리스트 항목을 불러와서 연결 리스트로 만들고
        node.next = prev # 노드의 다음부분에 연결리스트를 연결해줍니다.
        prev = node # 현재 노드를 저장해둡니다.

    return node

# 두 연결 리스트의 덧셈
def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
    a = self.toList(self.reverseList(l1)) # a와 b는 각각 l1, l2의 역순 리스트 입니다.
    b = self.toList(self.reverseList(l2)) # 설명에 제시된 샘플로 표현하면 각각 [3, 2, 1], [7, 6, 5]가 됩니다.

    resultStr = int(''.join(str(e) for e in a)) + \ # 설명에 제시된 샘플로 표현하면 resultStr은 321 + 765 = 1086 이 됩니다.
                int(''.join(str(e) for e in b))

    # 최종 계산 결과 연결 리스트 변환
    return self.toReversedLinkedList(str(resultStr)) # resultStr을 역순으로 연결리스트를 생성해줍니다. 최종 결과물은 6 -> 8 -> 0 -> 1이 됩니다.
```
