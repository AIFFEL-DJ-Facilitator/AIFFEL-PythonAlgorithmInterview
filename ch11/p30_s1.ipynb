```
def func(s):
    used = {} # (문자 : 마지막 인덱스) 형태로 저장됩니다.
    max_length = 0
    start = 0
    for index, char in enumerate(s):
        if char in used and start <= used[char]: # 
            start = used[char] + 1
            print(index, char, "start : ", start)
        else: # max_length는 
            max_length = max(max_length , index - start + 1)
            print(index, char, "max_length : ", max_length)

        used[char] = index
        print(used)
    return max_length
```

문자열이 abcabcbbc 일때, 첫 a, b, c는 used에 등록됩니다. {'a': 0, 'b': 1, 'c': 2} 또한 max_length는 3이 됩니다.
